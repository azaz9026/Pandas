```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = {
    'A':[1,2,3],
    'B':[4,5,6],
    'C':[7,8,9],
    'D':[10,20,30]
}

df = pd.DataFrame(data)

df
```

```python
df.plot()
plt.show()
```

```python
df['A'].plot(kind = 'line')
```

```python
df['B'].plot(kind = 'bar')
```

```python
df['B'].plot(kind = 'hist')
```

```python
df.plot( x = 'A' , y = 'B' , kind = 'scatter')
```

```python
plt.figure(figsize=(10,6))
df['A'].plot(kind='line' , linestyle = '--' , marker = '*' , color = 'red')
```

```python
df['B'].plot(kind='bar' , color = 'red')

plt.title('Value of B')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```
